```python
def generateText(source_text, max_len=100):
  res = generator(source_text, max_length=max_len, do_sample=True, temperature=0.9) 
  return res[0]['generated_text']
```

```python
sub_prompts = getSubpromptsWeb(prompt)
sub_prompts_text = []
for each_sub_prompt in sub_prompts:
    text = generateText(each_sub_prompt)
    sub_prompts_text.append({"sub_prompt_title":each_sub_prompt, "text":text})

```
